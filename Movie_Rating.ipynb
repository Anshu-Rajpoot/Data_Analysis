```python
import pandas as pd
df=pd.read_csv("movies.csv")
```

Explore the dataset

```python
# display the first few rows by default 5
print(df.head())
print(df.columns)
```

```python
#summarise stats
print(df.describe())
print(df.info())
```

```python
#check missing values
print(df.isna().sum())
```

Data cleaning

```python
#remove duplicates
df.drop_duplicates(inplace=True)
```

EDA (Exploratory Data Analysis)

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Distribution of ratings
plt.figure(figsize=(10,6))
sns.histplot(df['rating'], bins=10, kde=True)
plt.title("Movie Ratings")
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#Top 10 movies according to rating
top_movies=df.nlargest(10,"rating")
plt.figure(figsize=(15,6))
sns.barplot(x='title',y='rating',data=top_movies)
plt.title=('Top 10 movies according to Ratings')
plt.xticks(rotation=90)
plt.show()
```

```python
#Rating by Genre
plt.figure(figsize=(15,6))
sns.boxplot(x='genre',y='rating', data=df)
plt.xlabel("Genre")
plt.ylabel("Rating")
plt.xticks(rotation=90)
plt.show()
```
